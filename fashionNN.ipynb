So, what is neural network? It is the network of many mathematical models connected with each other through logical functions. # RBM?

A restricted Boltzmann machine (RBM) is a generative stochastic artificial neural network that can learn a probability distribution over its set of inputs. The restricted word in RBM means that there is no connection between the neurons the layer itself. Only connections areallowed are the connections between neurons of the two different layers, not withing the layer. We can see in the Fig1 RBM that the visible layers units have connections only with units of hidden layer and vice versa. The red units in visible layer are only connected with blue units in hidden layer.
<br>
<br>
In a minimal perspective, a neural network can be as small as containing only three layers. These layers in sequence from left to right are called --- Input Layer, Hidden Layer and Output Layer. 
<br>
<br>
The Input layer is a group of neurons that work as input and start of the neural network and feeds the input to the rest of the layers. THe Hidden layers are group of neurons that takes inputs from input layer or another hidden layer and supplies output to output layer or another hidden layer. Output layer recieves the final output. 
<br>
<br>
THe neurons are called nodes and the connect between the nodes are called connections. THese connections are nothing but weights given to preceding inputs and they are fed into the node of the hidden layer. In the nodes, the activation fucntion processes the inputs and passes on the output further.

# What is Activation Function?

<br>
<br>

### Three Layer Sequential Neural Network Model
<br>
<br>
After brief introduction of Tensorflow and Keras, let us begin with building the neural network model. We are building the sequential NN model. In such models, Layers of network are algined sequentially or stacking on each other like layers. FOllowing process flow will take place in model building 
<br>
<br>
INPUT LAYER >>>>>>>>>>  HIDDEN LAYER >>>>>>>>>>> OUTPUT LAYER
<br>
<br>
First layer will be <strong>Input Layer</strong>. In this case study, we will feed an array of 784 inputs as input layer. The 784 inputs are generated from standardized pixel values of 28 X 28 images. Each input of input layer will be fully connected with all the nodes of hidden layers. The weightage of each input will be determined in th model building. Similarly, After processing of inputs of input layer, activation fucntion of hidden layer will full connect with each variant of output array in output layer. The regression weights are basically connecting the output of hidden layer with the output layer. The output layer has an array often outcomes coded as 0 to 9. 
<br>
<br>

Let us see how we build the THREE LAYER SEQUENTIAL NEURAL NET MODEL with KERAS and TENSORFLOW.

#### INPUT LAYER
<br>
<br>
<strong>Input layer</strong> is created by flattening the 28 X 28 array of standardaized pixel values and it is passed on by the command "keras.layers.Flatten(input_shape=(28,28))".
<br>
<br>
#### HIDDEN LAYER
<br>
<br>
<strong>Hidden Layer</strong> is created as a layer of 128 nodes/neurons with activation function "reLU". The Hidden Layer gets input from Input Layer and regression weights are determined.The hidden layer is created by the command "keras.layers.Dense(128,activation="relu") 
<br>
<br>
#### OUTPUT LAYER
<br>
<br>
<strong>Output Layer</strong> contains array of 10 outputs and it is introduced by command "keras.layers.Dense(10,activation="softmax".  
<br>
<br>

```python
## Creating the Tensorflow Model

model1=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,activation="relu"),
    keras.layers.Dense(10,activation='softmax')
    ])

model1
```

### NN MODEL OPTIMIZATION
<br>
<br>
Every model is supposed to do prediction and it is obvious that there will be errors in prediction. Suppose, an image of bag is fed in the model, it should give the output as "Bag" but it may commit error. TO reduce the errors of the model, models needs to be optimized. Optimization means that the regression weights of the connections should be ontained in such a manner that it should reduce the error as mimimum as possible. As our case study model is prepared to identify the object and give it a right label, we are actually doing classification of the image. An image can have 10 possible outcomes. 
<br>
<br>
To reduce the error, an optimizer algorithm supervises the model development and tracks the classification errors. Here we are using "adam" as an optimizer and the target error is "sparse_categorical_crossentropy" and mesure of accuracy is "Accuracy". When error/loss decreases, the accuracy increases. This entire phase is called as Model optimization of Model Compilation. Follwing command chunk passes the optimzer, loss and metrics to the model---

```python

model1.compile(optimizer='adam', loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

### MODEL RERUNS & TESTING THE ACCURACY
<br>
<br>
Once model and compilation is ready, it is time to run the model and test its accuracy. Here "model1'fit" command runs the model 50 times (epocsh=10) on train dataset. After this, model is evaluated on test data to find how accurate the model is. 

```python
model1.fit(train_images, train_labels, epochs=50)

test_loss, test_acc=model1.evaluate(test_images,test_labels)

print("Tested Acc:", test_acc)
```

So, our final Test Accuracy is 88.68% on Three Layer Sewquential Model with 50 epochs. Now let's do the predictions and comare it with actual labels of the image.

### Predictions

Prediction is shown in terms of array of probabilties for each type of output. In our case, there are 10 outputs, so there will be array of 10 probabilties corresponding to each type of output. The probabilties are not comprehensible hence the outputtype which has got maximum probability will be chosen as the predicted label. 

```python
### Prediction with the model

prediction=model1.predict(test_images)

print(prediction)

```

```python
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array, true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  predictions_array, true_label = predictions_array, true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')

num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, prediction[i], test_labels, test_images)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i, prediction[i], test_labels)
plt.tight_layout()
plt.show()
```
